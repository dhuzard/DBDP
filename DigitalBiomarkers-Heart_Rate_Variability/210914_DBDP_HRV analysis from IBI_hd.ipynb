```python
import BIL_HRV_from_dataframe as bh
import pandas as pd
from datetime import datetime
import csv
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# PLOT THE IBI

# headers = ['time','IBI']
df = pd.read_csv('P10-ECG_All_IBI_data.csv', ";")

df = df.rename(columns={'time (s)':'time'})

print("5 first rows:")
print(df.head())
print("\n 5 last rows:")
print(df.tail())      
```

```python
#CONVERTS THE COMMA INTO A DOT AND MAKES IBI FLOATS:
df['IBI'] = df['IBI'].str.replace(',', '.').astype(float)
```

```python
filenames = df['file'].unique()

for i in filenames:
    
# df['time'] = df['time'].map(lambda x: datetime.strptime(str(x), '%Y/%m/%d %H:%M:%S.%f'))
    print(i)
    dfi = df.loc[df['file'] == i]
    dfi = dfi[['time', 'IBI']]
    print(dfi.head())
    
    #print(df.loc[df.file==i, 'Session'].values)
    x = dfi['time'].values
    y = dfi['IBI'].values

    # plot
    plt.plot(x,y)
    plt.show()
    
    print(bh.hrv_df(dfi))
```
